```python
import pandas as pd

# Especifica la ruta al archivo Excel.
ruta_archivo = 'base.xlsx'

# Lee el archivo Excel usando pandas.
try:
    df = pd.read_excel(ruta_archivo)

    # Convertir la columna 'Time Stamp' a formato datetime
    df['Time Stamp'] = pd.to_datetime(df['Time Stamp'])

    # Filtrar para el medidor 'T2_TGN_AA'
    filtro_medidor = df['MEDIDOR'] == 'T2_TGN_AA'
    df_filtrado = df[filtro_medidor]

    # Filtrar para el año 2024
    df_2024 = df_filtrado[df_filtrado['Time Stamp'].dt.year == 2024].copy()

    if not df_2024.empty:
        # Ordenar por Time Stamp
        df_2024 = df_2024.sort_values(by='Time Stamp')

        # Agrupar por mes y calcular el primer y último valor de SUM
        monthly_consumption = df_2024.groupby(df_2024['Time Stamp'].dt.month).agg(
            primer_valor=('SUM', 'first'),
            ultimo_valor=('SUM', 'last')
        )

        # Calcular el consumo mensual
        monthly_consumption['consumo_mensual'] = monthly_consumption['ultimo_valor'] - monthly_consumption['primer_valor']

        print(f"\nConsumo mensual del medidor T2_TGN_AA en el año 2024:")
        # Mapear el número del mes a su nombre
        month_names = {
            1: 'Enero', 2: 'Febrero', 3: 'Marzo', 4: 'Abril', 5: 'Mayo', 6: 'Junio',
            7: 'Julio', 8: 'Agosto', 9: 'Septiembre', 10: 'Octubre', 11: 'Noviembre', 12: 'Diciembre'
        }

        for month_number, row in monthly_consumption.iterrows():
            month_name = month_names.get(month_number, f'Mes {month_number}')
            print(f"{month_name}: {row['consumo_mensual']}")

    else:
        print("No se encontraron datos para el medidor T2_TGN_AA en el año 2024.")

except FileNotFoundError:
    print(f"Error: No se encontró el archivo '{ruta_archivo}'. Asegúrate de que el archivo exista y la ruta sea correcta.")
except KeyError as e:
    print(f"Error: No se encontró la columna '{e}'. Asegúrate de que los nombres de las columnas en tu archivo Excel coincidan con los nombres en el código ('MEDIDOR', 'Time Stamp', 'SUM').")
except Exception as e:
    print(f"Ocurrió un error al leer el archivo Excel: {e}")
```

```python
import pandas as pd

# Especifica la ruta al archivo Excel.
ruta_archivo = 'base.xlsx'

# Lee el archivo Excel usando pandas.
try:
    df = pd.read_excel(ruta_archivo)

    # Convertir la columna 'Time Stamp' a formato datetime
    df['Time Stamp'] = pd.to_datetime(df['Time Stamp'])

    # Lista de medidores para el edificio T2
    medidores_t2 = ['T2_TGN_AA', 'T2_TGN_INT']

    monthly_data = []

    for year in [2024]:
        for month in range(1, 13):
            month_data = {'Mes': f'{month:02d}/{year}'}
            total_consumo_mes = 0

            for medidor in medidores_t2:
                # Filtrar para el medidor actual y el año
                df_filtrado = df[(df['MEDIDOR'] == medidor) & (df['Time Stamp'].dt.year == year)]

                if not df_filtrado.empty:
                    # Obtener la última fecha del mes actual
                    ultimo_dia_mes = pd.Timestamp(year, month, 1) + pd.tseries.offsets.MonthEnd(0)
                    timestamp_fin_periodo = pd.to_datetime(f'{ultimo_dia_mes.year}-{ultimo_dia_mes.month:02d}-{ultimo_dia_mes.day} 00:00:00')
                    df_fin_periodo = df_filtrado[df_filtrado['Time Stamp'] == timestamp_fin_periodo]
                    valor_fin_periodo = df_fin_periodo['SUM'].max() if not df_fin_periodo.empty else None

                    # Obtener la última fecha del mes anterior
                    if month == 1:
                        mes_anterior = 12
                        anio_anterior = year - 1
                    else:
                        mes_anterior = month - 1
                        anio_anterior = year

                    ultimo_dia_mes_anterior = pd.Timestamp(anio_anterior, mes_anterior, 1) + pd.tseries.offsets.MonthEnd(0)
                    timestamp_fin_periodo_anterior = pd.to_datetime(f'{ultimo_dia_mes_anterior.year}-{ultimo_dia_mes_anterior.month:02d}-{ultimo_dia_mes_anterior.day} 00:00:00')
                    df_fin_periodo_anterior = df[(df['MEDIDOR'] == medidor) & (df['Time Stamp'].dt.year == anio_anterior) & (df['Time Stamp'].dt.month == mes_anterior) & (df['Time Stamp'].dt.date == ultimo_dia_mes_anterior.date()) & (df['Time Stamp'].dt.hour == 0) & (df['Time Stamp'].dt.minute == 0) & (df['Time Stamp'].dt.second == 0)]

                    valor_fin_periodo_anterior = df_fin_periodo_anterior['SUM'].max() if not df_fin_periodo_anterior.empty else None

                    consumo_mensual = None
                    if valor_fin_periodo is not None and valor_fin_periodo_anterior is not None:
                        consumo_mensual = valor_fin_periodo - valor_fin_periodo_anterior

                    month_data[f'Valor Fin Periodo ({medidor})'] = valor_fin_periodo
                    month_data[f'Timestamp Fin Periodo ({medidor})'] = timestamp_fin_periodo
                    month_data[f'Valor Fin Periodo Anterior ({medidor})'] = valor_fin_periodo_anterior
                    month_data[f'Timestamp Fin Periodo Anterior ({medidor})'] = timestamp_fin_periodo_anterior
                    month_data[f'Consumo Mensual ({medidor})'] = consumo_mensual
                    if consumo_mensual is not None:
                        total_consumo_mes += consumo_mensual

                else:
                    month_data[f'Valor Fin Periodo ({medidor})'] = None
                    month_data[f'Timestamp Fin Periodo ({medidor})'] = None
                    month_data[f'Valor Fin Periodo Anterior ({medidor})'] = None
                    month_data[f'Timestamp Fin Periodo Anterior ({medidor})'] = None
                    month_data[f'Consumo Mensual ({medidor})'] = None

            month_data['Consumo Total (T2)'] = total_consumo_mes if total_consumo_mes > 0 else None
            monthly_data.append(month_data)

    if monthly_data:
        df_tabla = pd.DataFrame(monthly_data)
        # Formatear las columnas de timestamp para mostrar la hora
        for medidor in medidores_t2:
            timestamp_col_actual = f'Timestamp Fin Periodo ({medidor})'
            timestamp_col_anterior = f'Timestamp Fin Periodo Anterior ({medidor})'
            if timestamp_col_actual in df_tabla.columns:
                df_tabla[timestamp_col_actual] = df_tabla[timestamp_col_actual].dt.strftime('%Y-%m-%d %H:%M:%S')
            if timestamp_col_anterior in df_tabla.columns:
                df_tabla[timestamp_col_anterior] = df_tabla[timestamp_col_anterior].dt.strftime('%Y-%m-%d %H:%M:%S')

        print("\nTabla de Consumo Mensual del Edificio T2 en el año 2024 (según lógica Power BI con timestamps a las 00:00):")
        print(df_tabla)
    else:
        print("No se encontraron datos de consumo para el edificio T2 en el año 2024.")

except FileNotFoundError:
    print(f"Error: No se encontró el archivo '{ruta_archivo}'. Asegúrate de que el archivo exista y la ruta sea correcta.")
except KeyError as e:
    print(f"Error: No se encontró la columna '{e}'. Asegúrate de que los nombres de las columnas en tu archivo Excel coincidan con los nombres en el código ('MEDIDOR', 'Time Stamp', 'SUM').")
except Exception as e:
    print(f"Ocurrió un error al leer el archivo Excel: {e}")
```
